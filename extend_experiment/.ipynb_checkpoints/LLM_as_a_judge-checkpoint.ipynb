```python
import pandas as pd
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

# Load your CSV
df = pd.read_csv("gen_text_replication_1000.csv")

# Initialize model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.to(torch.device("cuda" if torch.cuda.is_available() else "cpu"))
model.eval()

# Scoring function
def evaluate_text(text):
    prompt = (
        f"Evaluate the following text based on its semantic coherence and logical flow. "
        f"Provide only a score from 0 to 5.\nText: {text}\nScore:"
    )
    
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_length=inputs["input_ids"].shape[1] + 10,
            num_return_sequences=1,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )
    
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    try:
        score_str = generated_text.strip().split("Score:")[-1].strip().split()[0]
        score = int(score_str)
    except:
        score = -1  # fallback in case parsing fails
    return score

# Evaluate only a few rows for testing
df['coherence_score'] = df['text'].apply(evaluate_text)

# Group and calculate means by scaling factor
mean_scores = df.groupby('scaling_factor')['coherence_score'].mean().reset_index()

# Save
mean_scores.to_csv("mean_coherence_scores_by_scaling_factor.csv", index=False)
print(mean_scores)
```

```python
import re
import pandas as pd
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

# Load your CSV
df = pd.read_csv("gen_text_female_1000.csv")

# Initialize model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.to(torch.device("cuda" if torch.cuda.is_available() else "cpu"))
model.eval()

def evaluate_text(text):
    prompt = (
        f"Evaluate the following text based on its semantic coherence and logical flow. "
        f"Provide only a score from 0 to 5.\nText: {text}\nScore:"
    )
    
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_length=inputs["input_ids"].shape[1] + 20,
            num_return_sequences=1,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )
    
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    
    # Use regex to extract the first float number
    match = re.search(r"Score[:\s]*([0-5](?:\.\d+)?)", generated_text)
    if match:
        score = float(match.group(1))
    else:
        score = -1  # fallback for invalid output
    return score
# Evaluate only a few rows for testing
df['coherence_score'] = df['text'].apply(evaluate_text)

# Group and calculate means by scaling factor
mean_scores = df.groupby('scaling_factor')['coherence_score'].mean().reset_index()

# Save
mean_scores.to_csv("mean_coherence_scores_by_scaling_factor.csv", index=False)
print(mean_scores)
```

```python
import matplotlib.pyplot as plt

# Data
labels = [
    "-0.3 black + -0.3 fem",
    "0.0 black + 0.0 fem",
    "-0.3 black + 0.5 fem",
    "0.5 black + -0.3 fem",
    "0.5 black + 0.5 fem"
]
scores = [4.985, 4.484, 4.325, 4.401, 4.916]

# Plot
plt.figure(figsize=(10, 5))
bars = plt.bar(labels, scores, color='skyblue')
plt.ylabel('Mean Coherence Score')
plt.title('Coherence Score by Vector Combination')
plt.ylim(4.2, 5.1)

# Annotate bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.3f}', ha='center', va='bottom')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
