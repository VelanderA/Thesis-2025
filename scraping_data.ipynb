```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from bs4 import BeautifulSoup
import requests
import time

def get_rendered_html(url):
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(url)
    
    # Try to accept cookies if prompted
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".accept-btn"))
        )
        accept_button = driver.find_element(By.CSS_SELECTOR, ".accept-btn")
        accept_button.click()
    except TimeoutException:
        print("No cookie banner found or timeout expired.")
   
    # Click "Visa mer" until all content is loaded
    while True:
        try:
            show_more_button = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(text(),'Visa mer')]"))
            )
            driver.execute_script("arguments[0].scrollIntoView();", show_more_button)  # Scroll to button
            time.sleep(1)  # Allow scrolling to take effect
            show_more_button.click()
            print("Clicked 'Visa mer' button.")
            time.sleep(3)  # Allow content to load
        except (TimeoutException, NoSuchElementException):
            print("No more 'Visa mer' button found or all content is loaded.")
            break  # Exit loop if button is not found
   
    
    # Wait for the page to fully render
    time.sleep(5)  
    rendered_html = driver.page_source
    driver.quit()
    return rendered_html

def scrape_access_stories(url):
    rendered_html = get_rendered_html(url)
    soup = BeautifulSoup(rendered_html, 'html.parser')

    subpage_links = []
    for link in soup.find_all('a', href=True):
        subpage_url = link['href']
        if "/customer-cases/" in subpage_url:
            if not subpage_url.startswith('http'):
                subpage_url = requests.compat.urljoin(url, subpage_url)
            subpage_links.append(subpage_url)
    
    return subpage_links

# Updated base URL
url = "https://www.assaabloy.com/se/sv/stories/customer-cases"

# Scraping articles from the new URL
all_links = scrape_access_stories(url)

print(f"Total articles found: {len(all_links)}")
print("All article links:")
for link in all_links:
    print(link)
```

```python
"""anonymize with syntetic data"""

import spacy
import json
from faker import Faker

# Load Swedish spaCy model
nlp = spacy.load("sv_core_news_lg")
fake = Faker("sv_SE")  # Swedish locale

# Dictionary to store mappings
entity_map = {}

def get_synthetic_name(entity, label):
    """Generate or reuse synthetic names for detected entities."""
    if entity not in entity_map:
        if label == "ORG":
            synthetic_name = fake.company()
        elif label == "PRS":
            synthetic_name = fake.name()
        else:
            return entity  
        
        entity_map[entity] = synthetic_name
    
    return entity_map[entity]

# Load data
with open("merged_assa_deduplicated_paragraphs.json", "r", encoding="utf-8") as f:
    data = json.load(f)

synthetic_data = []

# Process each text entry
for entry in data:
    if isinstance(entry, str):
        doc = nlp(entry)
        new_text = entry
        for ent in doc.ents:
            if ent.label_ in ["PRS", "ORG"]:
                synthetic_name = get_synthetic_name(ent.text, ent.label_)
                new_text = new_text.replace(ent.text, synthetic_name)
        synthetic_data.append(new_text)

# Save the result
with open("synthetic_data.json", "w", encoding="utf-8") as f:
    json.dump(synthetic_data, f, ensure_ascii=False, indent=4)

print("Synthetic data with Faker has been saved.")
print(f"Entities mapped: {len(entity_map)}")
print(entity_map)
```

```python
import spacy
import json
import re
from faker import Faker

# Load the Swedish spaCy model
nlp = spacy.load("sv_core_news_lg")
fake = Faker("sv_SE")  # Swedish locale

# Load file
with open('merged_assa_deduplicated_paragraphs.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Combine all text into one string
full_text = " ".join(data)

# Process full text with spaCy
doc = nlp(full_text)

# Create a copy of the original text
anonymized_text = full_text

# Dictionary to map entities to synthetic names
entity_map = {}

# Detect named entities with spaCy
for ent in doc.ents:
    if ent.label_ in ["PERSON", "ORG"]:
        # Generate synthetic data if not already in the mapping
        if ent.text not in entity_map:
            fake_name = fake.name() if ent.label_ == "PERSON" else fake.company()
            entity_map[ent.text] = fake_name

        # Replace with same fake name
        anonymized_text = anonymized_text.replace(ent.text, entity_map[ent.text])

# Regex to catch missed names
regex_names = re.findall(r'\b[A-ZÅÄÖ][a-zåäö]+\b', full_text)

# Filter out already processed names and duplicates
regex_names = list(set(regex_names) - set(entity_map.keys()))

# Save anonymized text
with open('Banonymized_data.json', 'w', encoding='utf-8') as f:
    json.dump({"anonymized_text": anonymized_text}, f, ensure_ascii=False, indent=4)

# Save entity mapping
with open('Bentity_mapping.json', 'w', encoding='utf-8') as f:
    json.dump(entity_map, f, ensure_ascii=False, indent=4)

# Save regex_names
with open('regex_names.json', 'w', encoding='utf-8') as f:
    json.dump(regex_names, f, ensure_ascii=False, indent=4)
    
print("Anonymization complete.")

```
